```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data = {17.5: [15, 18, 19, 20], 18: [16, 17, 18, 19], 18.5: [14, 16, 17], 19: [15, 16], 20: [13, 14, 15], 20.5: [12, 13, 14], 21: [10, 11, 13, 15], 21.5: [10, 12, 13], 22: [8, 10, 11, 12], 23: [8, 9, 10], 23.5: [7, 8, 9, 11], 24: [6, 8, 9], 24.5: [6, 7, 8, 10], 25: [4, 6, 7, 8], 25.5: [5, 6], 26: [4, 5, 7], 26.5: [3, 4, 5], 27: [2, 3, 4, 7], 27.5: [2, 3], 28: [0, 1, 2, 4], 28.5: [0, 2], 29: [0, 1, 2]}
feet_size = []
number_of_mistakes = []
for x, y in data.items():
    for mistake in y:
        feet_size.append(x)
        number_of_mistakes.append(mistake)
        
full_data = pd.DataFrame({"feet size": feet_size, "number of mistakes": number_of_mistakes})

feet_size = [x for x in data.keys()]
avg_number_of_mistakes = [np.average(x) for x in data.values()]
avg_data = pd.DataFrame({"feet size": feet_size, "average number of mistakes": avg_number_of_mistakes})
```

```python
plt.figure(figsize=(20, 12))
plot = sns.scatterplot(full_data["feet size"], full_data["number of mistakes"])
sns.lineplot(avg_data["feet size"], avg_data["average number of mistakes"], color="green")
plot.set_xticks(np.arange(17, 29.5, 0.5))
plt.show()
```
